```python
data['YEAR'].value_counts().sort_index()
data['DISTRIBUTOR'].value_counts().head(20)
```

```python
streaming = ['Netflix','Hulu','Prime Video','Disney+','']
data["platform_type"] = data["DISTRIBUTOR"].apply
(lambda x: "Streaming" if x in streaming else "Traditional")
```

```python
nominations_by_year = (
    data.groupby(["YEAR", "platform_type"])
        .size()
        .reset_index(name="Nominations"))
```

```python
import matplotlib.pyplot as plt

top_distributors = (
    data["DISTRIBUTOR"]
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10,6))
top_distributors.sort_values().plot(kind="barh")

plt.title("Top 10 Television Distributors by Emmy Nominations")
plt.xlabel("Number of Nominations")
plt.ylabel("Distributor")

plt.tight_layout()
plt.show()
```
